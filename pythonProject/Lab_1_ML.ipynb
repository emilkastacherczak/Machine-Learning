```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.datasets import load_iris
# load the famous iris data
irisRaw = load_iris()
```

```python
#1 Please view the data and refer to the specific information.

# read iris.data into a pandas DataFrame (df), including column names
iris = pd.DataFrame(data= np.c_[irisRaw['data'], irisRaw['target']],
                     columns= irisRaw['feature_names'] + ['target'])
iris
```

```python
#2 Display the number of rows and columns.

iris.shape
```

```python
#3 View basic information for individual columns. Use the 'describe' and 'groupby' methods

print(iris.describe())

iris.groupby('sepal length (cm)').mean()  #mean values for every sepal length value
```

```python
#4 Using the 'head' method please display the first 5 rows

iris.head(5)
```

```python
#5 Using “dropna” check if the database contains missing data

iris.dropna()
```

```python
#6 Sort the data in ascending order relative to the second column

iris.sort_values('sepal width (cm)')
```

```python
#7 Find the minimum and maximum petal length (column 3 - petal length). Give the index of each of these values.

print('Minimum petal length (cm): ', iris["petal length (cm)"].min(), ' Index: ', iris["petal length (cm)"].idxmin())
print('Maximum petal length (cm): ', iris.iloc[:, 2].max(), ' Index: ', iris.iloc[:, 2].idxmax())
```

```python
#8 Calculate the standard deviation for each column

iris.std()
```

```python
#9 Identify rows for which the length of the flower cup (sepal length) is greater than the average length (the whole set)

iris[ iris['sepal length (cm)'] > iris['sepal length (cm)'].mean() ]
```

```python
#10 Display histogram

plt.hist(iris['sepal length (cm)'])
plt.xlabel('sepal length (cm)')
plt.ylabel('Quantity')
plt.show()
plt.hist(iris['sepal width (cm)'])
plt.xlabel('sepal width (cm)')
plt.ylabel('Quantity')
plt.show()
plt.hist(iris['petal length (cm)'])
plt.xlabel('petal length (cm)')
plt.ylabel('Quantity')
plt.show()
plt.hist(iris['petal width (cm)'])
plt.xlabel('petal width (cm)')
plt.ylabel('Quantity')
plt.show()
```
